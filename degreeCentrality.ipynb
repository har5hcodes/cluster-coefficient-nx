```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

G = nx.Graph()

edge_list=  [('A', 'D'),('A', 'E'),('A', 'F'),
            ('B', 'C'),('B', 'D'),('B', 'E'),('B', 'F'),
            ('C', 'D'),('C', 'Q'),('C', 'O'),
            ('D', 'E'),('D', 'Q'),
            ('E', 'Q'),('F', 'Q'),
            ('Q', 'O') ]

G.add_edges_from(edge_list)

nx.draw(G, with_labels = True)

print(nx.degree_centrality(G))


plt.show()
```
