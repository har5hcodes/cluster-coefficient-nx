```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

edge_list=  [('A', 'C'),
            ('B', 'C'),('B', 'D'),('B', 'E'),
            ('C', 'D'),('C', 'H'),
            ('D', 'E'),('D', 'H'),
            ('E', 'F'),('E', 'G')]

G.add_edges_from(edge_list)

nx.draw(G, with_labels = True)

print(nx.clustering(G, 'C'))
print(nx.clustering(G, 'D'))

plt.show()
```
